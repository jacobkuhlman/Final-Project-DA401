**1. Functions that will be used throughout the notebook**

```python
def openData(path):
    with open(path) as f:
        data = json.load(f)
    return data 


def rename_column(oldName, newName, df):
    """This function renames columns of a pandas dataframe. 
    Parameters:
        oldName: Name of the old column in which you want to change 
        newName: Text in which you want the column to be renamed
        oldName2: Name of the old column in which you want to change 
        newName2: Text in which you want the column to be renamed
        df: The dataframe you want to rename columns in. 
    
    Returns a dataframe with new column names. 
    """
    New = df.rename(columns={oldName: newName})
    return New

def get_year(df):
    df["year"] = df['created_at'].astype(str).str[:4]
    return df

def drop(df, column):
    df = df.drop([column], axis=1)
    return df 

def get_month(df):
    df['Month'] = df['created_at'].str[5:7]
    return df

def get_day(df):
    df['Day'] = df['created_at'].str[8:10]
    return df

def DayTweeted(df):
    df['Date'] = df['Date Tweeted'].str[0:10]
    return df
```
